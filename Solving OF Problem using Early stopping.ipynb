```python
from sklearn.datasets import make_moons
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
```

```python
X,y = make_moons(n_samples=100,noise =0.2, random_state=1) #generating the dummy data with 100 rows , generating IV (X) DV (y)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
X_train,X_test = X[:30,:],X[30:,:]
```

```python
y_train,y_test = y[:30],y[30:]
```

```python
model = Sequential() #brain without neurons
```

```python
model.add(Dense(500, input_dim = 2, activation = 'relu')) #input
```

```python
model.add(Dense(1, activation = 'sigmoid')) #o/player
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(mode='min',monitor='val_loss', patience=200)
```

```python
history = model.fit(X_train,y_train, validation_data = (X_test,y_test), epochs = 4000, callbacks=[es])
```

```python
plt.plot(history.history['loss'], c='b', label = 'Training Loss')
plt.plot(history.history['val_loss'], c='r', label = 'Testing Loss')
```
